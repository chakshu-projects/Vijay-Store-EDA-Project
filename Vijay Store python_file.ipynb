#### 7). Exploring Data & Visualization

```python
# Total revenue 
total_revenue = df['Amount'].sum()
print(total_revenue.astype('int'))
```

```python
# revenue by month
Monthly_sales = df.groupby('Month')['Amount'].sum().astype('int')
print(Monthly_sales.sort_values(ascending = False))
```

```python
plt.figure(figsize = (15,5))
Monthly_sales.sort_values(ascending = False).plot(kind='line',marker='o')
plt.title('MONTHLY SALES')
plt.xlabel('MONTH')
plt.ylabel('REVENUE')
```

#### Observation: The month January has the highest sales and August has the lowest sales.

```python
# top product by revenues
top_product = df.groupby('product_name')['Amount'].sum().head(5).astype('int')
top_5_prod = pd.DataFrame(top_product.sort_values(ascending = False))
top_5_prod
```

```python
plt.figure(figsize = (15,5))
sns.barplot(x='product_name',y='Amount',data=top_5_prod)
plt.title('TOP 5 PRODUCTS BY REVENUE')
plt.xlabel('PRODUCT NAME')
plt.ylabel('REVENUE')
```

#### Observation: Dell Inspiron and Laptop Bag are the top revenue-generating products, contributing highest sales among all products shown.

```python
# Top customers
top_customer = df.groupby('customer_id')['Amount'].sum().head(5).astype('int')
top_5_cust = pd.DataFrame(top_customer.sort_values(ascending = False))
top_5_cust
```

```python
plt.figure(figsize = (15,5))
sns.barplot(x='customer_id',y='Amount',data=top_5_cust)
plt.title('TOP 5 Customer')
plt.xlabel('CUSTOMER_ID')
plt.ylabel('REVENUE')
```

#### Observation: Customer C0001 generated the highest revenue, contributing around 1.08 million, making it the top-performing customer.

```python
# Total Revenue by city
city_sales = df.groupby('city')['Amount'].sum().astype('int')
print(city_sales.sort_values(ascending = False))
```

```python
plt.figure(figsize = (15,5))
city_sales.sort_values(ascending = False).plot(kind='line',marker='o')
plt.title('sales of City')
plt.xlabel('city')
plt.ylabel('REVENUE')
```

#### Observation: The charts shows an overall downward trend in revenue across cities from Delhi to Chennai.

```python
# Quantity by months
Monthly = df.groupby('Month')['quantity'].sum()
print(Monthly.sort_values(ascending = False))
```

```python
plt.figure(figsize = (15,5))
Monthly.sort_values(ascending = False).plot(kind='line',marker='o')
plt.title('MONTHLY Quantity')
plt.xlabel('MONTH')
plt.ylabel('Quantity')
```

#### Observation: The chart suggest that product quantity sold decreased consistently from the beginning to the end of the year.

```python
# Top 5 product by Quantity
top_5_p = df.groupby('product_name')['quantity'].sum().head(5)
top_product = pd.DataFrame(top_5_p.sort_values(ascending = False))
top_product
```

```python
plt.figure(figsize = (15,5))
sns.barplot(x='product_name',y='quantity',data=top_product)
plt.title('TOP 5 PRODUCTS BY Quantity')
plt.xlabel('PRODUCT NAME')
plt.ylabel('Quantity')
```

#### Observation: Dell Inspiron has the highest quantity sold, with around 360 units, making it the most in-demand product.

```python
# Total Quantity by Year
Year_qu = df.groupby('Year')['quantity'].sum()
year_quantity = pd.DataFrame(Year_qu)
print(Year_qu.sort_values(ascending = False))
```

```python
plt.figure(figsize = (15,5))
sns.barplot(x='Year',y='quantity',data=year_quantity,)
plt.title('Quantity by Year')
plt.xlabel('Year')
plt.ylabel('Quantity')
```

#### Observation: The quantity in 2024 is significantly higher than in 2025, showing that most sales/orders occurred in 2024.

```python
# top 5 cities based on Quantity
Top_city = df.groupby('city')['quantity'].sum().head(5)
top_5_ct = pd.DataFrame(Top_city.sort_values(ascending = False))
top_5_ct
```

```python
plt.figure(figsize = (15,5))
sns.barplot(x='city',y='quantity',data=top_5_ct)
plt.title('TOP 5 CITIES BY Quantity')
plt.xlabel('City')
plt.ylabel('Quantity')
```

#### Observation: There is a significant gap between Delhi and the other cities, suggesting that Delhi contributes nearly double the quantity of Bengaluru and Hyderabad.

```python
numeric_df = df.select_dtypes(np.number)
sns.heatmap(numeric_df.corr(),annot=True)
plt.title('Correlation')
```

#### Observation: 
  -  Quantity and Amount show a high positive correlation (0.78), this means when the quantity increases, the total amount also increases significantly.
  -  Unit Price and Base Price have a moderate positive correlation (0.54), suggesting that products with higher base prices generally also have higher selling prices.
  -  Small negative correlations such as between Quantity and Base Price (-0.049) indicate that higher-priced items may be bought slightly less often, but the effect is extremely weak      

####  Key Insights
  -  The month January has contributed a major part in revenue and quantity solds.
  -  The product Dell Inspiron is the top performing among all the other products.
  -  Most part of the sales and quantity sold are take place in Delhi.
  -  The correlation graph help the business to understand the relation between the data.
